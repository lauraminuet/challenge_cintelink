### Bases de productos


```python
df = data_meli.obtener_info_de_mercado_libre()
```

```python
df.head()
```

```python
df_destino = pd.DataFrame(columns=['id', 'title','price', 'sold_quantity','thumbnail'])
```

```python
for index, row in df.iterrows():
    df_destino = df_destino.append(row['results'], ignore_index=True)
```

```python
df_destino.head()
```

```python
df_final = pd.DataFrame()

df_final[['id', 'title','price', 'sold_quantity','thumbnail']] = df_destino[['id', 'title','price', 'sold_quantity','thumbnail']]
```

```python
df_final.head()
```

```python
df_final.shape
```

```python
df_final['site_id'] = df_final['id'].str.slice(stop=3)
df_final['created_date'] = datetime.today().date()
```

```python
df_final.head()
```

```python
engine = create_engine('sqlite:///productos_meli.sqlite3', echo=True)
```

```python
df_final.to_sql('productos', con=engine, if_exists="replace")
```
